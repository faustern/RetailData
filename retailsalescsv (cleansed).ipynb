```python
df["Year"] = df["Order Date"].dt.year
df
```

```python
Gain_orders_2009 = df[(df["Profit"] > 1) & (df["Year"] == 2009)]
Loss_orders_2009 = df[(df["Profit"] < 1) & (df["Year"] == 2009)]
Gain_order_count_2009 = Gain_orders_2009["Order ID"].nunique()
Loss_order_count_2009 = Loss_orders_2009["Order ID"].nunique()
print(f"Số đơn hàng lãi trong năm 2009: {Gain_order_count_2009}")
print(f"Số đơn hàng lỗ trong năm 2009: {Loss_order_count_2009}")
```

```python
Gain_orders_2012 = df[(df["Profit"] > 1) & (df["Year"] == 2012)]
Loss_orders_2012 = df[(df["Profit"] < 1) & (df["Year"] == 2012)]
Gain_order_count_2012 = Gain_orders_2012["Order ID"].nunique()
Loss_order_count_2012 = Loss_orders_2012["Order ID"].nunique()
print(f"Số đơn hàng lãi trong năm 2012: {Gain_order_count_2012}")
print(f"Số đơn hàng lỗ trong năm 2012: {Loss_order_count_2012}")
```

```python
Order = {
    "Year": [2009, 2010, 2011, 2012],
    "Gain Order": [Gain_order_count_2009, Gain_order_count_2010, Gain_order_count_2011, Gain_order_count_2012],
    "Loss Order": [Loss_order_count_2009, Loss_order_count_2010, Loss_order_count_2011, Loss_order_count_2012]
}
Order_Status = pd.DataFrame(Order)
Order_Status
```
